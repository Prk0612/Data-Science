```python
import pytesseract
import cv2
from pdf2image import convert_from_path
from PIL import Image
import numpy as np
import os

# Set the TESSDATA_PREFIX environment variable
os.environ['TESSDATA_PREFIX'] = r'C:\Users\Priyank\AppData\Local\Programs\Python\Python310\Lib\site-packages\Tesseract-OCR\tessdata'

# PDF to Image Conversion
def pdf_to_images(pdf_path):
    images = convert_from_path(pdf_path, dpi=600)  # Adjust the DPI value as needed for desired image resolution
    image_paths = []
    for i, image in enumerate(images):
        image_path = f'output_image_{i}.png'  # Output image path with PNG extension
        image.save(image_path, 'PNG')  # Save the image in PNG format
        image_paths.append(image_path)
    return image_paths

# Load the PDF
pdf_path = r'D:\Freelancer\SEC PDF extraction\Part2_314571.pdf'  # Path to input PDF file
image_paths = pdf_to_images(pdf_path)

# Process each image separately
for i, image_path in enumerate(image_paths):

    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply adaptive thresholding to create a mask for background
    _, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours in the mask
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a white background image
    result = np.ones_like(image) * 255

    # Draw filled contours (background) in black
    cv2.drawContours(result, contours, -1, (0, 0, 0), thickness=cv2.FILLED)

    # Invert the mask
    mask_inv = cv2.bitwise_not(mask)

    # Apply the inverted mask to the original image to preserve text
    result = cv2.bitwise_and(image, image, mask=mask_inv)

    # Extract text using Tesseract
    text = pytesseract.image_to_string(image, config='--psm 6')

    # Display the result
    print('Extracted Text:')
    print(text)

```
