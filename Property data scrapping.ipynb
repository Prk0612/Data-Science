```python
from bs4 import BeautifulSoup
import requests
from google.oauth2 import service_account
from googleapiclient.discovery import build
from time import sleep as wait

SERVICE_ACCOUNT_FILE = 'key.json'

SCOPES = ['https://www.googleapis.com/auth/spreadsheets']
# creds = None
creds = service_account.Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)
# The ID and range of a sample spreadsheet.
SAMPLE_SPREADSHEET_ID = '1NF9fcu0eeZF6T9cXfke1SskSDky8XdzwqXX969YmhJI'
service = build('sheets', 'v4', credentials=creds)
# Call the Sheets API
sheet = service.spreadsheets()
result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID,
                            range='Format Kosan 2!A3:AF').execute()
h = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 '
                  'Safari/537.36 Edg/113.0.1774.42'}

ad_list = []
num = 35
no = 0
for i in range(1, num):
    url = 'https://www.lamudi.co.id/buy/?page=' + str(i)
    response = requests.get(url, headers=h).text
    soup = BeautifulSoup(response, 'html.parser')
    print(f'Preparing:{url}')
    ad_link = soup.find_all('div', 'ListingCell-AllInfo ListingUnit')
    link = [i.find_next('a').get('href') for i in ad_link]
    appender = [ad_list.append(i) for i in link]

print(f'Total Pages Ready: {len(ad_list)}, the results should be {(num - 1) * 30}')
print(f'Launching Deep Scrape In Progress...')

sheet.values().clear(spreadsheetId=SAMPLE_SPREADSHEET_ID, range='Format Kosan 2!A3:AF').execute()
for u in ad_list:
    r = requests.get(u, headers=h).text
    s = BeautifulSoup(r, 'html.parser')
    try:
        no += 1
        title = s.find('h1', 'Title-pdp-title').get_text().strip()
        print(title)
        price = float(s.find('span', 'FirstPrice').get_text().replace('Rp', '').replace('.', '').strip())
        city = s.find('p', 'Location').get_text().strip()
        broker_name = s.find('div', 'AgentInfoV2-agent-name').get_text().strip()
        broke_phone = s.find('div',
                             'AgentInfoV2-requestPhoneSection-phoneNumber AgentInfoV2-requestPhoneSection-gradient').get_text().strip()
    except:
        title = None
        price = None
        city = None
        broker_name = None
        broke_phone = None

    try:
        rooms = int(s.select_one('div[data-attr-name="bedrooms"]').find_next_sibling('div', 'last').get_text().strip())
    except:
        rooms = None

    try:
        ac = s.select_one('div[data-slug="air-conditioning"]').find_next('span',
                                                                         'listing-amenities-name').get_text().strip().replace(
            'Pendingin ruangan (AC)', 'Yes')
    except:
        ac = None

    try:
        parking = s.select_one('div[data-slug="secure-parking"]').find_next('span',
                                                                            'listing-amenities-name').get_text().strip().replace(
            'Secure parking', 'Yes')
    except:
        parking = None
    try:
        land_area = s.select_one('div[data-attr-name="land_size"]').find_next_sibling('div', 'last').get_text().strip()
    except:
        land_area = None
    try:
        bathroom = s.select_one('div[data-attr-name="bathrooms"]').find_next_sibling('div', 'last').get_text().strip()
    except:
        bathroom = None
    try:
        building = s.select_one('div[data-attr-name="building_size"]').find_next_sibling('div',
                                                                                         'last').get_text().strip()
    except:
        building = None
    try:
        secure_parking = s.select_one('div[data-slug="secure-parking"]').find_next('span',
                                                                                   'listing-amenities-name').get_text().replace(
            'Secure parking', 'Yes')
    except:
        secure_parking = None
    try:
        balcony = s.select_one('div[data-slug="balcony"]').find_next('span',
                                                                     'listing-amenities-name').get_text().replace(
            'Balkon', 'Yes')
    except:
        balcony = None
    try:
        wifi = s.select_one('div[data-slug="broadband-internet-available"]').find_next('span',
                                                                                       'listing-amenities-name').get_text().replace(
            'Internet broadband/Wifi', '1200000').replace('Internet broadband','1200000')
    except:
        wifi = None
    try:
        tv_cable = s.select_one('div[data-slug="pay-tv-access"]').find_next('span',
                                                                            'listing-amenities-name').get_text().replace(
            'TV kabel', '40000')
    except:
        tv_cable = None
    try:
        ad_date = s.select_one('div[data-attr-name="empty"]').find_next_sibling('div', 'last').get_text()
    except:
        ad_date = None
    # data-slug="gym"
    try:
        deck = s.select_one('div[data-slug="deck"]').find_next('span', 'listing-amenities-name').get_text().replace(
            'Dek', 'Yes')
    except:
        deck = None
    try:
        gym = s.select_one('div[data-slug="gym"]').find_next('span', 'listing-amenities-name').get_text().replace('Gym',
                                                                                                                  'Yes')
    except:
        gym = None
    try:
        swimming_pool = s.select_one('div[data-slug="swimming-pool"]').find_next('span',
                                                                                 'lising-amenities-name').get_text().replace(
            'Kolam renang', 'Yes')
    except:
        swimming_pool = None
    try:
        park = s.select_one('div[data-slug="garden"]').find_next('span', 'lising-amenities-name').get_text().replace(
            'Taman', 'Yes')
    except:
        park = None

    try:
        kitchen_set = s.select_one('div[data-slug="kitchen-set"]').find_next('span',
                                                                             'lising-amenities-name').get_text().replace(
            'Kitchen set', 'Yes')
    except:
        kitchen_set = None
    try:
        security = s.select_one('div[data-slug="247-security"]').find_next('span',
                                                                           'lising-amenities-name').get_text().replace(
            'Keamanan 24 jam', '1530000')
    except:
        security = None
    try:
        water_heater = s.select_one('div[data-slug="hot-water"]').find_next('span',
                                                                            'lising-amenities-name').get_text().replace(
            'Air Panas', '400000')
    except:
        water_heater = None

    try:
        loan = float(s.select_one('#loan-calculator-v2-container > section > div > div.Loan-Summary > div > table > tbody > tr:nth-child(2) > td:nth-child(2)').get_text().replace('INR',''))
    except:
        loan = None
    finally:
        gen_list = [[
            no,
            "Yousef",
            title,
            "Kosan",
            u,
            "Indonesia",
            city,
            rooms,
            land_area,
            building,
            bathroom,
            ad_date,
            broker_name,
            broke_phone,
            price,
            None,
            wifi,
            tv_cable,
            water_heater,
            security,
            None,  # floors,
            ac,
            parking,
            kitchen_set,
            gym,
            None,  # ent_area,
            park,
            secure_parking,
            swimming_pool,
            balcony,
            deck,
        ]]

        print(f'Sending: {title}')

    # sending to google sheets through API
    request = sheet.values().append(spreadsheetId=SAMPLE_SPREADSHEET_ID, range='Format Kosan 2!A3:AF',
                                    valueInputOption='USER_ENTERED', body={'values': gen_list}).execute()
    print(f'Successfully Scrapped: {title}')
    wait(0.45)
```
